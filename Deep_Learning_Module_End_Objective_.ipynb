```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_diabetes
from sklearn.preprocessing import StandardScaler
```

```python
diabetes = load_diabetes()#LOAD DATA SET

X = diabetes.data
y = diabetes.target
```

```python
df = pd.DataFrame(X, columns=diabetes.feature_names)
df["target"] = y

df.head()
```

```python
print(df.shape)
```

```python
df.info()
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
X = df.drop("target", axis=1)
y = df["target"]#Separate Features and Target
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
X_scaled = pd.DataFrame(X_scaled, columns=diabetes.feature_names)

X_scaled.head()
```

```python
df.describe()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Build ANN model
model = Sequential([
    Dense(64, activation='relu', input_shape=(10,)),  # Hidden Layer 1
    Dense(32, activation='relu'),                     # Hidden Layer 2
    Dense(1, activation='linear')                     # Output Layer
])

# Display model summary
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,      # or X if you didn't scale the data
    y,
    test_size=0.2,
    random_state=42
)
history = model.fit(
    X_train,
    y_train,
    validation_data=(X_test, y_test),
    epochs=100,
    batch_size=32,
    verbose=1
)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Evaluate the model
test_loss, test_mae = model.evaluate(X_test, y_test)

print("Test Loss (MSE):", test_loss)
print("Test MAE:", test_mae)
```

```python
# Predict using ANN model
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)

print("Mean Squared Error:", mse)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)

print("R² Score:", r2)
```

Improved ANN Model Code

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

improved_model = Sequential([
    Dense(128, activation='relu', input_shape=(10,)),
    Dropout(0.2),

    Dense(64, activation='relu'),

    Dense(32, activation='relu'),

    Dense(1, activation='linear')
])

improved_model.summary()
```

```python
improved_model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)
```

```python
history_improved = improved_model.fit(
    X_train,
    y_train,
    validation_data=(X_test, y_test),
    epochs=150,
    batch_size=16,
    verbose=1
)
```

```python
from sklearn.metrics import mean_squared_error, r2_score #Evaluate Improved Model

# Predictions
y_pred_improved = improved_model.predict(X_test)

# Metrics
mse_improved = mean_squared_error(y_test, y_pred_improved)
r2_improved = r2_score(y_test, y_pred_improved)

print("Improved MSE:", mse_improved)
print("Improved R2 Score:", r2_improved)
```

The ANN model was improved by increasing the number of neurons, adding an additional hidden layer, introducing dropout regularization, and increasing the number of training epochs. The improved model showed better performance compared to the baseline model, with reduced Mean Squared Error and increased R² score. Dropout helped reduce overfitting, while additional neurons allowed the network to learn more complex patterns from the diabetes dataset.

```python
from tensorflow.keras.callbacks import EarlyStopping # ADD EARLY STOPPING

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)

history = improved_model.fit(
    X_train,
    y_train,
    validation_data=(X_test,y_test),
    epochs=200,
    callbacks=[early_stop]
)
```
